## Analysis and Optimization Checklist Discussion

The design variables in this analysis are the disc inner diameter, outer diameter, and thickness. All of the variables in this problem are continuous. The constraints include the yield strength and melting point of the brake disc material as well as the firing frequency of the engine. The objectives include minimizing the volume, minimizing the maximum stress, maximizing the first natural frequency, and minimizing the maximum temperature in the brake disc.

There are several potential trade-offs between the objectives of the problem, considered below.
- The first and most obvious is that minimization of the volume involves minimization of material used, and this can reduce the amount of stress that can be handled by the brake disc.
- Similiarly, the minimization of material can reduce the stiffness of the brake disc, and this may cause a reduction in the maximum natural frequency of the disc.
- The heat transfer of the brake disc due to convection will increase with larger surface area, but heat transfer by conduction within the disc will be less effective with more material. Increasing the surface area while decreasing the volume (large diameter and small thickness) is the best case for the thermal analysis and will minimize the maximum temperature in the disc.

The objective and constraint functions in this problem are based on analytical methods but ANSYS uses numerical approximations.

Based on the above discussion and due to having multiple objectives, the Adaptive Multi-Objective Optimization (AMO) algorithm in ANSYS will be used to find the optimal solution.

A Design of Experiments (DoE) was performed using ten (10) samples of variable combinations and Latin Hypercube Sampling. The results are shown below.

It is clear that Sample 10 has the lowest volume, and, as predicted, experiences the highest stress and maximum temperature. However, these values are within the operational constraints of the material used for the brake disc, and therefore, Sample 10 demonstrates the optimal combination of inner diameter, outer diameter, and thickness of the brake disc. There is clearly infeasibility in the total deformation due to the natural frequency of the brake disc and these results should be ignored until the cause of this is determined.

Unfortunately, the Adaptive Multi-Objective Optimization (AMO) did not produce optimal results, and there may be issues with the design of the optimization problem in ANSYS. Further analysis would be necessary to determine the issue with this optimization analysis.
